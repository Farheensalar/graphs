```python
import matplotlib.pyplot as plt
```

```python
plt.plot([1,1,1,1],[1,2,3,4])
plt.plot([4,4,4,4],[1,2,3,4])
plt.plot([1,2,3,4],[1,1,1,1])
plt.plot([1,2,3,4],[4,4,4,4])
plt.plot([1.5,1.5,1.5,1.5],[1.5,2,3,3.5])
plt.plot([3.5,3.5,3.5,3.5],[1.5,2,3,3.5])
plt.plot([1.5,2,3,3.5],[1.5,1.5,1.5,1.5])
plt.plot([1.5,2,3,3.5],[3.5,3.5,3.5,3.5])
plt.show()
```

```python
plt.plot([1,2,3,4],[4,8,14,6], 'g--' ,marker='*',mec='b',lw=3)
plt.xlabel('x-Axis')
plt.ylabel('y-Axis')
plt.title('Demo-Line chart')
plt.show()
```

```python
x=[10,40,60,80,90,52,100,25,52,45,56,65,12,15,17,18,25,35,15,15,15]
y=[10,20,30,40,60,80,100]
plt.xlabel('x-Axis')
plt.ylabel('y-Axis')
plt.hist(x,y)
plt.show()
```
